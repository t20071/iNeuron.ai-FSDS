```python
dataset['Date'] = dataset['Date_of_Journey'].str.split(('/')).str[0]
dataset['Month'] = dataset['Date_of_Journey'].str.split(('/')).str[1]
dataset['Year'] = dataset['Date_of_Journey'].str.split(('/')).str[2]
```

Data,month and year still the object

```python
dataset['Date'] = dataset['Date'].astype(int)
dataset['Month'] = dataset['Month'].astype(int)
dataset['Year'] = dataset['Year'].astype(int)
```

```python
dataset.info()
```

```python
dataset.drop('Date_of_Journey',inplace = True,axis = 1)
```

```python
dataset.info()
```

```python
# find out the duration of flight
dataset['new_Arrival_Time']=dataset['Arrival_Time'].str.split(' ').str[0] #remove the month name and create new column
```

```python
# create arrival hour table
dataset['Arrival_hour'] = dataset['new_Arrival_Time'].apply(lambda x:x.split(":")[0])  

# create arrival minute table 
dataset['Arrival_minute'] = dataset['new_Arrival_Time'].apply(lambda x:x.split(":")[1])
```

```python
dataset.drop(['Arrival_Time','new_Arrival_Time'],inplace = True,axis = 1) # remove the both the column 
```

```python
dataset.info()
```

> arrival hour and minute are in object

```python
dataset['Arrival_hour'] = dataset['Arrival_hour'].astype(int)
dataset['Arrival_minute'] = dataset['Arrival_minute'].astype(int)
```

```python
dataset.info()
```

```python
dataset.head()
```

```python
# same for the departure time 
dataset['Dept_hour'] = dataset['Dep_Time'].str.split(':').str[0]
dataset['Dept_min'] = dataset['Dep_Time'].str.split(':').str[1]
dataset['Dept_hour'] = dataset['Dept_hour'].astype(int)
dataset['Dept_min'] = dataset['Dept_min'].astype(int)
dataset.drop('Dep_Time',axis=1,inplace= True)
```
