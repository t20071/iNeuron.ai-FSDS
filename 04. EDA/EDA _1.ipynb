```python
# find out missing values
dataset.isnull().sum()
```

```python
[features for features in dataset.columns] # gives the column
```

```python
features_with_na = [features for features in dataset.columns if dataset[features].isnull().sum()>=1] # gives the column
features_with_na # features with nan value
```

```python
dataset['Electrical'].isnull().mean()
```

```python
# % of null value in dataset['features']
for feature in features_with_na:
    print(feature,np.round(dataset[feature].isnull().mean()*100,4),'% missing value')
```

```python
sns.boxplot(dataset['SalePrice'])
```

> So many outlier ie why median take in the below code

```python
# lets plot some amazing the diagrams 
# where nan put 1, and there is value put 0, then plot graph
# then group by 1 and 0 and see the output
data = dataset.copy()
for feature in features_with_na:
    
    # lets make a variable that indicate 1 if the obeseravatin was missing
    data[feature] = np.where(data[feature].isnull(),1,0)
    
    # lets calculte the mean SalePrice where the information is missing or present
    data.groupby(feature)['SalePrice'].median().plot.bar()
    plt.title(feature)
    plt.show()
```

Here with the relation between the missing value and dependant variable is clearly visible.So, we need to replace these nan value with something meaningfull which we will do in the feature engg section.

### Temporal variable (Eg - Datetime variables)

```python
dataset.info()
```

```python
year_feature = [feature for feature in numerical_features if 'Yr' in feature or 'Year' in feature]
year_feature
```

```python
for feature in year_feature:
    print(feature,dataset[feature].unique())
```

```python
dataset.groupby('YrSold')['SalePrice'].median().plot()
plt.xlabel('Year sold')
plt.ylabel('Median house price')
plt.title('House price vs year sold')
plt.show()
```

In US, recession during 2008

```python
# Here, we will compare the differance bw all year feature with salesprice 
data = dataset.copy()
for feature in year_feature:
    if feature != 'YrSold':
        data[feature] = data['YrSold'] - data[feature]
        plt.scatter(data[feature],data['SalePrice'])
        plt.xlabel(feature)
        plt.ylabel('SalePrice')
        plt.show()
```

 ### Observation : 
 Within all the year information, it is clearly visible that the new houses are more costiler that the older one, looks like a pareto or power distribution. 

## Numerical variable 
1. Continous variable
1. Discrete variable

```python
discrete_feature = [feature for feature in numerical_features if len(dataset[feature].unique())<=25]
# 25 decide by the domain expert
print(len(discrete_feature))
discrete_feature
```

```python
dataset[discrete_feature].head()
```

```python
# find the relationship bw discrete and saleprice
data = dataset.copy()
for feature in discrete_feature:
    data.groupby(feature)['SalePrice'].median().plot.bar()
    plt.xlabel(feature)
    plt.ylabel('SalePrice')
    plt.title(feature)
    plt.show()
```

### observation 
There is relationship bw discrete variable and saleprice

```python
## continous variable 
continous_feature = [feature for feature in numerical_features if feature not in discrete_feature]
print(len(continous_feature))
continous_feature
```

```python
### lets analyse the continous value by creating histogram to understand the distribution 
data = dataset.copy()
for feature in continous_feature:
     data.groupby(feature)['SalePrice'].median().plot.hist()
     plt.xlabel(feature)
     plt.ylabel('SalePrice')
     plt.title(feature)
     plt.show()
```

```python
data = dataset.copy()
for feature in continous_feature:
    data[feature].hist(bins=25)
    plt.xlabel(feature)
    plt.ylabel('count')
    plt.title(feature)
    plt.show()
```

# EDA PART 2

```python
# we will be using log transformation
data = dataset.copy()
for feature in continous_feature:
    if 0 in data[feature].unique():
        pass
    else:
        data[feature]=np.log(data[feature])
        plt.scatter(data[feature],data['SalePrice'])
        plt.xlabel(feature)
        plt.ylabel('SalePrice')
        plt.title(feature)
        plt.show()
```

## Outlier

```python
for feature in continous_feature:
    if 0 in data[feature].unique():
        pass
    else:
        data[feature] = np.log(data[feature])
        data.boxplot(column = feature)
        plt.ylabel(feature)
        plt.title(feature)
        plt.show()
        
```
