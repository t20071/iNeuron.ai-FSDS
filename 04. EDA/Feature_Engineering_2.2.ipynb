### Numerical variable_with nan value  

```python
numerical_features_nan = [feature for feature in dataset.columns if dataset[feature].dtype != 'O' 
                      and dataset[feature].isnull().sum()>=1]
numerical_features_nan
```

```python
for feature in numerical_features_nan:
    
    # since there are outlier we are going to replace with median 
    median_values = dataset[feature].median()
    
    ## create a new feature to capture nan value
    dataset[feature+' nan'] = np.where(dataset[feature].isnull(),1,0)
    
    ## replace nan in feature with median value only in numerical_feature so there will not be nan value in feature 
    dataset[feature].fillna(median_values,inplace = True)
```
